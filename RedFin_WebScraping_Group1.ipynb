```python
#getting alameda county webpage links
redfin_listings = []

#alameda county
for x in range(1, 9, 1):
    base_url = "https://www.redfin.com/county/303/CA/Alameda-County"
    page = "/page-" + str(x) 
    base_url = base_url + page
    redfin_listings.append((base_url))
    
#san francisco county
for x in range(1, 9, 1):
    base_url = "https://www.redfin.com/county/340/CA/San-Francisco-County"
    page = "/page-" + str(x) 
    base_url = base_url + page
    redfin_listings.append((base_url))
    
#santa clara county
for x in range(1, 9, 1):
    base_url = "https://www.redfin.com/county/345/CA/Santa-Clara-County"
    page = "/page-" + str(x) 
    base_url = base_url + page
    redfin_listings.append((base_url))
    
#contra costa county
for x in range(1, 9, 1):
    base_url = "https://www.redfin.com/county/345/CA/Santa-Clara-County"
    page = "/page-" + str(x) 
    base_url = base_url + page
    redfin_listings.append((base_url))
    
print(redfin_listings)
```

```python
#data cleaning
import pandas as pd
import numpy as np


#read csv to clean
df=pd.read_csv('Data_RedFin_Raw.csv',index_col=0)

#replacing expressions in data with nan
x = df.columns
df[x] = df[x].replace('—', np.nan, regex = True)
#df[x] = df[x].replace('-', np.nan, regex = True)
df[x] = df[x].replace('', np.nan, regex = True)

#Dropping old index
#df = df.drop(df.columns[0], axis = 1)

#Dropping any sqft rows without a proper value in them
#df = df[df['sqft'] != 'â€”']
#df.drop(df[df['sqft']=="Äî'"].index,inplace=True)
#df.drop(df[df['sqft']=='â€”'].index,inplace=True)

#rounding floats to two places
df['bath'] = df['bath'].astype(float)  #as bath is non int type
df["bath"] = df["bath"].round(2)

df['1st School Rating'] = df['1st School Rating'].round(2)
df["Lotsize"] = df["Lotsize"].round(2)
df["perSq_Ft"] = df["perSq_Ft"].round(2)

# function for lotsize
# def lot_size (row):
#     if float(row['Lotsize']) == 0:
#         return row['sqft']
#     return row['Lotsize']


#Changing Squarefootage to an Int
#df['sqft'] = df['sqft'].applymap(lambda x: str(x.replace(',',"")))
df['sqft'] = df['sqft'].str.replace(',',"")
#df['sqft'] = df['sqft'].str.replace(',','').astype(float)
#df['sqft'] = df['sqft'][:5].astype(float)

#renaming columns
df.rename(columns = {"county": "city"}, 
          inplace = True)

#adding new price level column
df['Price_Level'] = [ 0 if index <= 1000000
                             else 1
                             for index in df['price']]

#removing url column
df.drop('url',axis=1,inplace=True)

#removing duplicates
df.drop_duplicates()

#df['bed']=df[['bed']].applymap(lambda x: x.encode('utf-8').decode('ascii', 'ignore'))
#df['bath']=df[['bath']].applymap(lambda x: x.encode('utf-8').decode('ascii', 'ignore'))
df['bed'].replace(' ', np.nan, inplace=True) #converting blanks to nan
df.dropna(subset=['bed'], inplace=True)     #dropping nan values
df['bath'].replace(' ', np.nan, inplace=True) #converting blanks to nan
df.dropna(subset=['bath'], inplace=True) #dropping nan values

df['city']=df['city'].apply(lambda x:x[0]+x[1:].lower()) #normalizing values across all cities
df['state']=df['state'].apply(lambda x:x[0]+x[1:].lower()) #normalizing values across all cities

df['walkscore']=df['walkscore'].astype(int) # converting walkscore to integer

df['price']=df['price'].astype(int)# converting price to integer


print(len(df))

```

```python
#inserting median value by city for any missing quantitative values

for col in ['price', 'bed', 'bath', 'sqft',
       'walkscore', 'transitscore', 'bikescore', 'competitivescore',
       '1st School Rating', 'year_built', 'Lotsize',
       'perSq_Ft']:
    df[col] = df[col].fillna(df.groupby('city')[col].transform('median'))
```

```python
#Removing outliers for select columns
#Did not want to remove for all columns because found that it affects the data visualizations heavily

#year_built
Q1 = df['year_built'].quantile(0.25)
Q3 = df['year_built'].quantile(0.75)
IQR = Q3 - Q1

df = df[(df['year_built'] >= (Q1 - 1.5 * IQR)) & (df['year_built'] <= (Q3 + 1.5 * IQR))]
```

```python
#drop all null and rest index
df = df.dropna()
df.reset_index(inplace=True, drop=True)
```
